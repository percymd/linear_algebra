# Matrices identidad inversa singular

```python
identidad = np.eye(4)
print(identidad)
```

```python
vector = np.array([[2],[3],[5],[7]])
print(vector)
```

```python
print(identidad.dot(vector))
```

```python
A = np.array([[1,0,1],[0,1,1],[-1,1,1]])
print(A)
```

```python
inversa_A = np.linalg.inv(A)
print(inversa_A)
```

```python
print(A.dot(inversa_A))
```

```python
#Cuando no existe la inversa, se llama matriz singular
singular = np.array([[1,1],[1,1]])
print(singular)
```

# Ejemplos de sistemas de ecuaciones lineales sin solucion

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-6,6)

y_1 = 3*x + 5
y_2 = -x + 3
y_3 = 2*x + 1
```

```python
plt.figure()
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.xlim(-8, 8)
plt.ylim(-8, 8)

plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')

plt.show()
```

Deben juntarse las 3 lineas. Al no juntarse se considera sobredeterminado, significa que se tiene mas ecuaciones que variables


### Ejemplo 2

```python
x = np.arange(-6, 6)

y_2 = -1*x + 3
y_3 = 2*x +1
```

```python
plt.figure()
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.xlim(-8, 8)
plt.ylim(-8, 8)

plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')

plt.show()
```

Igual de ecuaciones y variables es un sistema determinado. Si no se cruzan ambas rectas es un sistema incompatible.

### Ejemplo 3

```python
x = np.arange(-6, 6)

y_3 = 2*x +1
```

```python
plt.figure()
plt.plot(x, y_3)

plt.xlim(-8, 8)
plt.ylim(-8, 8)

plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')

plt.show()
```

Cuando se tiene una sola ecuacion con dos variable, nuestro sistema tiene un grado de libertad, cualquier valor de x es la solucion. Tiene soluciones infinitas
